```python
# Importing the required libraries
import seaborn as sns
import pandas as pd
```

```python
#  View the Datasets and loading one of them
sns.get_dataset_names()
data = sns.load_dataset('titanic')
data
```

```python
# Convert the dataset into a DataFrame
df = pd.DataFrame(data)
df
```

```python
# To get the null values
df.isnull().sum()
```

```python
# Catching column "age" of null values for better understanding
df['age'].head(50)
```

```python
# To fill the missing values with Means
mean_age = df['age'].mean()
df['age mean'] = df['age'].fillna(mean_age)
```

```python
# To fill the missing values with Median
median_age = df['age'].median()
df['age median'] = df['age'].fillna(median_age)
```

```python
df['deck'].head(10)
```

```python
# To fill the missing values with Mode
mode_deck = df['deck'].mode()[0]
df['deck mode'] = df['deck'].fillna(mode_deck)
```

```python
# Dropping the column
df_dropped = data.drop('deck mode',axis=1)
```

```python
# To get all column names
df.columns
```

```python
# Renaming the column "sex" by "Gender"
df_rename = df.rename(columns = {'sex':'Gender'})
df_rename.head()
```

```python
# Selecting the column with integer datatype
integer_data = df.select_dtypes('int')
integer_data
```

```python
# Grouping the column "who" by fare
df.groupby('who').agg({'fare':'count'})
```

```python
# Maps in Pandas
df['survived'] = df['survived'].map(lambda x : 'Survived' if x == 1 else 'Not Survived')
```

```python
# Replace
df['who'] = df['who'].replace(['man','woman'],['M','F'])
```

```python
# To save
df.to_csv('/content/sample_data/data1.csv',index = False)
```
